```python
from sklearn.metrics import mean_squared_error

def predict(input_data,model,labels):

    preds = model.predict(input_data)
    mse = mean_squared_error(labels,preds)
    rmse = np.sqrt(mse)
    rmse

    return rmse
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error, r2_score
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

def plot_importances(df, col_name, title, top_n=15):
  top = df.sort_values(col_name, ascending=False).head(top_n)
  plt.figure(figsize=(8, 0.5*len(top)+1))
  sns.barplot(x=col_name, y="Feature", data=top)
  plt.title(title)
  plt.show()

param_space = {"copy_X": [True, False], "fit_intercept": [True, False],
               "n_jobs": [1, 5, 10, 15, None], "positive": [True, False]}

war_lin = LinearRegression()
aav_lin = LinearRegression()

war_grid = GridSearchCV(war_lin, param_space, cv=5)
aav_grid = GridSearchCV(war_lin, param_space, cv=5)

war_grid.fit(war_x_train, war_y_train)
aav_grid.fit(aav_x_train, aav_y_train)

war_best = LinearRegression(**war_grid.best_params_)
aav_best = LinearRegression(**aav_grid.best_params_)

war_best.fit(war_x_train, war_y_train)
aav_best.fit(aav_x_train, aav_y_train)

war_pred = war_best.predict(war_x_test)
aav_pred = aav_best.predict(aav_x_test)

war_rmse = np.sqrt(mean_squared_error(war_y_test, war_pred))
aav_rmse = np.sqrt(mean_squared_error(aav_y_test, aav_pred))
war_r2 = r2_score(war_y_test, war_pred)
aav_r2 = r2_score(aav_y_test, aav_pred)
train_war_r2 = r2_score(war_y_train, war_best.predict(war_x_train))
train_aav_r2 = r2_score(aav_y_train, aav_best.predict(aav_x_train))

print(f"All Players WAR RMSE: {war_rmse:.4f}, R² Test: {war_r2:.4f}, R² Train: {train_war_r2:.4f}")
print(f"All Players AAV RMSE: {aav_rmse:.4f}, R² Test: {aav_r2:.4f}, R² Train: {train_aav_r2:.4f}")

war_feat = pd.DataFrame({'Feature': war_x_train.columns, 'fWAR Feature Importance': war_best.coef_})
aav_feat = pd.DataFrame({'Feature': aav_x_train.columns, 'AAV Feature Importance': aav_best.coef_})

print(war_feat.sort_values('fWAR Feature Importance', ascending=False).head(15))
print(aav_feat.sort_values('AAV Feature Importance', ascending=False).head(15))

plot_importances(war_feat, 'fWAR Feature Importance', "fWAR Feature Importance - All Players")
plot_importances(aav_feat, 'AAV Feature Importance', "AAV Feature Importance - All Players")

inf_war_lin = LinearRegression()
inf_aav_lin = LinearRegression()
inf_war_lin.fit(inf_war_x_train, inf_war_y_train)
inf_aav_lin.fit(inf_aav_x_train, inf_aav_y_train)
inf_war_pred = inf_war_lin.predict(inf_war_x_test)
inf_aav_pred = inf_aav_lin.predict(inf_aav_x_test)
inf_war_rmse = np.sqrt(mean_squared_error(inf_war_y_test, inf_war_pred))
inf_aav_rmse = np.sqrt(mean_squared_error(inf_aav_y_test, inf_aav_pred))
inf_war_r2 = r2_score(inf_war_y_test, inf_war_pred)
inf_aav_r2 = r2_score(inf_aav_y_test, inf_aav_pred)

train_inf_war_r2 = r2_score(inf_war_y_train, inf_war_lin.predict(inf_war_x_train))
train_inf_aav_r2 = r2_score(inf_aav_y_train, inf_aav_lin.predict(inf_aav_x_train))
print(f"Infielders WAR RMSE: {inf_war_rmse:.4f}, R² Test: {inf_war_r2:.4f}, R² Train: {train_inf_war_r2:.4f}")
print(f"Infielders AAV RMSE: {inf_aav_rmse:.4f}, R² Test: {inf_aav_r2:.4f}, R² Train: {train_inf_aav_r2:.4f}")

inf_war_feat = pd.DataFrame({'Feature': inf_war_x_train.columns, 'fWAR Feature Importance for Infielders': inf_war_lin.coef_})
inf_aav_feat = pd.DataFrame({'Feature': inf_aav_x_train.columns, 'AAV Feature Importance for Infielders': inf_aav_lin.coef_})
print(inf_war_feat.sort_values('fWAR Feature Importance for Infielders', ascending=False).head(15))
print(inf_aav_feat.sort_values('AAV Feature Importance for Infielders', ascending=False).head(15))
plot_importances(inf_war_feat, 'fWAR Feature Importance for Infielders', "fWAR Feature Importance - Infielders")
plot_importances(inf_aav_feat, 'AAV Feature Importance for Infielders', "AAV Feature Importance - Infielders")

of_war_lin = LinearRegression()
of_aav_lin = LinearRegression()
of_war_lin.fit(of_war_x_train, of_war_y_train)
of_aav_lin.fit(of_aav_x_train, of_aav_y_train)
of_war_pred = of_war_lin.predict(of_war_x_test)

of_aav_pred = of_aav_lin.predict(of_aav_x_test)
of_war_rmse = np.sqrt(mean_squared_error(of_war_y_test, of_war_pred))
of_aav_rmse = np.sqrt(mean_squared_error(of_aav_y_test, of_aav_pred))
of_war_r2 = r2_score(of_war_y_test, of_war_pred)
of_aav_r2 = r2_score(of_aav_y_test, of_aav_pred)
train_of_war_r2 = r2_score(of_war_y_train, of_war_lin.predict(of_war_x_train))
train_of_aav_r2 = r2_score(of_aav_y_train, of_aav_lin.predict(of_aav_x_train))
print(f"Outfielders WAR RMSE: {of_war_rmse:.4f}, R² Test: {of_war_r2:.4f}, R² Train: {train_of_war_r2:.4f}")
print(f"Outfielders AAV RMSE: {of_aav_rmse:.4f}, R² Test: {of_aav_r2:.4f}, R² Train: {train_of_aav_r2:.4f}")

of_war_feat = pd.DataFrame({'Feature': of_war_x_train.columns, 'fWAR Feature Importance for Outfielders': of_war_lin.coef_})
of_aav_feat = pd.DataFrame({'Feature': of_aav_x_train.columns, 'AAV Feature Importance for Outfielders': of_aav_lin.coef_})
print(of_war_feat.sort_values('fWAR Feature Importance for Outfielders', ascending=False).head(15))
print(of_aav_feat.sort_values('AAV Feature Importance for Outfielders', ascending=False).head(15))
plot_importances(of_war_feat, 'fWAR Feature Importance for Outfielders', "fWAR Feature Importance - Outfielders")
plot_importances(of_aav_feat, 'AAV Feature Importance for Outfielders', "AAV Feature Importance - Outfielders")
```
